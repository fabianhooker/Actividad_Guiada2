```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
import sympy as sp





```

```python



f= lambda X: X[0]**2 +X[1]**2
df= lambda X: [2*X[0], 2*X[1]]

resolucion = 30
rango=2.5
X=np.linspace(-rango,rango,resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))

for ix,x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[iy,ix]=f([x,y])

        
contour=plt.contourf(X,Y,Z,resolucion)
cbar=plt.colorbar(contour)

P=[random.uniform(-rango,rango),random.uniform(-rango,rango)]
plt.plot(P[0],P[1],"o",c="red")
TA=0.05

for _ in range(10):
    grad=df(P)
    P[0], P[1]=P[0]-TA*grad[0],P[1]-TA*grad[1]
    plt.plot(P[0],P[1],"o",c="red")
    
plt.plot(P[0],P[1],"o",c="green")
print("Solucion: ", P,f(P))

        
```

# problema 2 
f(x)=sin(1/2*x^2 - 1/4*y^2 + 3)*cos(2*x+1-e^y)


```python
# Definimos la funcin y su derivada
f = lambda x: np.sin(1/2 * x[0]*2 - 1/4 * x[1]*2 + 3) * np.cos(2 * x[0] + 1 - np.exp(x[1]))
df_dx = lambda x: x[0] * np.cos(2 * x[0] + 1 - np.exp(x[1])) * np.cos(1/2 * x[0]*2 - 1/4 * x[1]*2 + 3)
df_dy = lambda x: -0.25 * x[1] * np.sin(2 * x[0] + 1 - np.exp(x[1])) * np.cos(1/2 * x[0]*2 - 1/4 * x[1]*2 + 3)

resolucion = 50
rango = 0.005
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

for ix, x in enumerate(X):
    for iy, y in enumerate(Y):
        Z[iy, ix] = f([x, y])

contour = plt.contourf(X, Y, Z, resolucion)
cbar = plt.colorbar(contour)

P = [random.uniform(-rango, rango), random.uniform(-rango, rango)]
plt.plot(P[0], P[1], "o", c="red")
TA = 0.05

for _ in range(20):
    grad_x = df_dx(P)
    grad_y = df_dy(P)
    P[0], P[1] = P[0] - TA * grad_x, P[1] - TA * grad_y
    plt.plot(P[0], P[1], "o", c="red")

plt.plot(P[0], P[1], "o", c="green")
print("Solucion: ", P, f(P))
```
